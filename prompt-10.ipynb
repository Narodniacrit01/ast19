```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 0
std_dev = 1

random_numbers = np.random.normal(mean, std_dev, 1000)

plt.hist(random_numbers, bins=100, alpha=0.7, color='orange', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers from Normal Distribution')

plt.savefig('histogram_normal_distribution.pdf')
plt.show()
```
