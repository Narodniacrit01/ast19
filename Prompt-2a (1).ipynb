```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)

def sin_func(x):
    return np.sin(x)

def cos_func(x):
    return np.cos(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].plot(x, sin_func(x), color='Pink')
axs[0].set_title('sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

axs[1].plot(x, cos_func(x), color='green')
axs[1].set_title('cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

plt.tight_layout()
plt.savefig('trig_functions.pdf')
plt.show()
```
