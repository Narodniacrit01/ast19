# Sine and Cosine Functions

This notebook defines two functions: `sin_function` and `cos_function`. These functions compute the sine and cosine of an input value `x`, respectively. 

We will also create a table of values for `sin(x)` and `cos(x)` over the interval from 0 to 2 with a thousand entries, and finally, we will print out the first 10 values of `x`, `sin(x)`, and `cos(x)` in columns.

```python
#python cell
import numpy as np

def sin_function(x):
    #Computes the sine of x.
    #Parameters 
    #x : float or array-like
    #The input value(s) for which to compute the sine.
    #Returns: float or ndarray
    #The sine of the input value(s).
    return np.sin(x)
#Python cell
```

```python
#python cell
def cos_function(x):
    #Computes the cosine of x.
    #Parameters 
    #x : float or array-like
    #The input value(s) for which to compute the cosine.
    #Returns: float or ndarray
    #The cosine of the input value(s).
    return np.cos(x)
#Python cell
```

## Tabulating Sine and Cosine Values

In this section, we will create an array of `x` values between 0 and 2, with 1000 evenly spaced entries. We will then compute `sin(x)` and `cos(x)` using our defined functions and display the results in a table.

```python
#python cell
# Create an array of x values
x_values = np.linspace(0, 2, 1000)

# Compute sine and cosine values
sin_values = sin_function(x_values)
cos_values = cos_function(x_values)

# Tabulate results in a structured way
results_table = np.column_stack((x_values, sin_values, cos_values))
#python cell
```

## Displaying the First 10 Values

We will use a for loop to print the first 10 values of `x`, `sin(x)`, and `cos(x)` in a nicely formatted table.

```python
#python cell
# Print the first 10 values of x, sin(x), and cos(x)
print(f"{'x':<10} {'sin(x)':<10} {'cos(x)':<10}")
for i in range(10):
    print(f"{x_values[i]:<10.2f} {sin_values[i]:<10.6f} {cos_values[i]:<10.6f}")
```
