```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,2*3.141592653589793) 
y = np.sin(x)          
plt.plot(x,y)          
plt.xlabel('x')        
plt.ylabel('y')        
#plt.show()             
plt.savefig('sinx.png',bbox_inches='tight',dpi=1000)
```
